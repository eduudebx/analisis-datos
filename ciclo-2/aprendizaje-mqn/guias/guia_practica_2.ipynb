```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#df = pd.read_excel('data/BASE PARA ANALISIS-M2A.xlsx', sheet_name=0)
#df.to_csv('data/BASE PARA ANALISIS-M2A.csv', sep='|', index=False)
df = pd.read_csv('data/BASE PARA ANALISIS-M2A.csv', sep='|')
df.head(1)
```

```python
# Extraendo todas los nombres de las columnas para diccionario de datos en un archivo auxiliar.
with open('data/guia2_aux.txt', 'w') as f:
    for col in df.columns:
        f.write(f"{col}\n")

# Columnas y Filas del DataFrame
print(f'\n---- CSV ----\nNúmero de filas: {df.shape[0]}\nNúmero de Columnas: {df.shape[1]}\n')

# Leemos el archivo que contiene el diccionario de datos.
dic_datos = {}
with open('data/guia2_dic_datos.txt', 'r') as archivo:
    for linea in archivo:
        linea = linea.strip()
        if '|' in linea:
            clave, valor = linea.split('|', 1)
            dic_datos[clave] = valor

# Número de datos en el Diccionario.
print(f'\n---- DIC ----\nNúmero de datos: {len(dic_datos)}\n')

#with open('data/guia2_diccionario_datos.txt', 'r') as file:
    #claves = file.readline().strip().split(',')
    #valores = file.readline().strip().split(',')
    #diccionario_datos = dict(zip(claves, valores))
```

```python
nom_columnas = list(dic_datos.keys())
df.columns = nom_columnas
df.head(10)
```

```python
nulos_vacios = {}
for col in nom_columnas:
    nulos = df[col].isnull().sum()
    vacios = (df[col] == '').sum()

    if (nulos + vacios) > 0:
        nulos_vacios[col] = f'Nulos: {nulos} | Vacios: {vacios}'

print(f'\n---- INFO ----\nColumnas con valores nulos: {len(nulos_vacios)}\n\n')
for i in nulos_vacios:
    print(f'{i}: {nulos_vacios[i]}')
```
