<b>Importando los datos de un archivo excel en un dataframe de Pandas:</b>

```python
import pandas as pd


df = pd.read_excel('datos_estudiantes.xlsx', sheet_name=0)
df.head(10)
```

<hr>
<b>Identificando valores nulos por columna:</b>

```python
nulos_por_columna = df.isnull().sum()
print(f'\nNulos por columna:\n\n{nulos_por_columna}')
```

<hr>
<b>Obteniendo información del dataframe:</b>

```python
print(f'\nInformacion:\n\n{df.info()}')
```

<hr>
<b>En la columna 'Edad' se identifican cadenas de texto que no son números enteros por lo que se procede a reemplazar por el promedio de las edades:</b>

```python
df['Edad'] = df['Edad'].apply(lambda x: x if str(x).isnumeric() else '0')
df['Edad'] = df['Edad'].astype(int)

promedio = round(df[df['Edad'] != 0]['Edad'].mean())
print(f'\nPromedio de edades: {promedio}\n')

df['Edad'] = df['Edad'].replace(0, promedio)

df.head(10)
```

<hr>
<b>En las columnas 'Nota1' y 'Nota2' se identifican valores negativos, tal vez por un error de tipeo, por lo que se procede a reemplazar por el valor absoluto de la nota:</b>

```python
df['Nota1'] = df['Nota1'].astype(int).abs()
df['Nota2'] = df['Nota2'].astype(int).abs()

df.head(10)
```

<hr>
<b>La columna 'Asistencia' contiene un patrón de True y False, por lo que se procede a transformar la columna en sus valores booleanos correspondientes:</b>

```python
print(df['Asistencia'].unique())
df['Asistencia'] = df['Asistencia'].replace({'True': True, 'False': False})

df.head(10)
```

<hr>
<b>La columna 'Fecha_Inscripcion' contiene registros identificados como 'Error', por lo que se procede a cambiar a 'Desconocido':</b>

```python
df['Fecha_Inscripcion'] = pd.to_datetime(df['Fecha_Inscripcion'], errors='coerce')
df['Fecha_Inscripcion'] = df['Fecha_Inscripcion'].fillna('Desconocido')

df.head(10)
```

<hr>
<b>Exportamos los datos limpios en un nuevo archivo de Excel:</b>

```python
df.to_excel('datos_limpios_estudiantes.xlsx', index=False)
```
